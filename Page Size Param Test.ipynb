```python
import requests
import json
from pprint import pprint
import pandas as pd
```

```python
# Seeing total amount of records based on "total"
response = requests.get('https://api.fbi.gov/wanted/v1/list')
data = json.loads(response.content)
print(data['total'])
print(data['items'][0]['title'])
```

```python
# Testing parameters
response = requests.get('https://api.fbi.gov/wanted/v1/list', params={
'pageSize': 1, 'page': 1
})
data = json.loads(response.content)
print(data['page'])
print(data['items'][0]['title'])
    
```

```python
num = 1
titles = []
subjects = []

#Looping through all records. Set parament of pageSize so that the 'page' paramemter detemines the specific record
for x in range(924): 
    response = requests.get('https://api.fbi.gov/wanted/v1/list', params={
    'pageSize': 1, 'page': num
    })
    data = json.loads(response.content)
    titles.append(data['items'][0]['title'])
    subjects.append(data['items'][0]["subjects"])
    num +=1
```

```python
# Creating data frame based on whatever info we pull. 
fbi_dict = {
    "Title": titles,
    "Subject": subjects
}

fbi_df = pd.DataFrame(fbi_dict)
fbi_df
```
